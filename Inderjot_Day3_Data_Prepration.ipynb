```python
## We need to change data type of 8 attributes
```

```python
## i.e. ('Suburb','Address','Type','Method','SellerG','CouncilArea','Regionname')
## These 7 attributes are in 'object' type but these need to be in 'Category' type
```

```python
## 'Date' is in 'object' type it needs to in 'Date' type
```

```python
## Changing their Data Type
```

```python
melbourne['Suburb']=melbourne['Suburb'].astype('category')
melbourne['Address']=melbourne['Address'].astype('category')
melbourne['Type']=melbourne['Type'].astype('category')
melbourne['Method']=melbourne['Method'].astype('category')
melbourne['SellerG']=melbourne['SellerG'].astype('category')
melbourne['CouncilArea']=melbourne['CouncilArea'].astype('category')
melbourne['Regionname']=melbourne['Regionname'].astype('category')
melbourne['Date']=pd.to_datetime(melbourne['Date'])
```

```python
melbourne.info()
```

```python
pd.isna(melbourne).sum()
```

```python
## Price
##We'll replace the missing price values by median of price
```

```python
melbourne['Price'].fillna(melbourne.groupby('Type')['Price'].transform(lambda x : x.median()),inplace=True)
```

```python
## Distance,Postcode,Bedroom2,Bathroom,Car,Yearbuilt
##Landsize,BuildingArea,CounsilArea
##Lattitude,Longitude,Regionname,Propertycount
```

```python
##We'll replace all of these attributes missing values by their mode's
##As many of the houses follow same pattern(for eg:2BHK)
##There're other options also like(3BHK,4BHK or Penthouses)
##But most of the houses similar pattern
```

```python
distance_mode = melbourne['Distance'].mode()
postcode_mode = melbourne['Postcode'].mode()
bedroom_mode = melbourne['Bedroom2'].mode()
bathroom_mode = melbourne['Bathroom'].mode()
car_mode = melbourne['Car'].mode()
yearbuilt_mode = melbourne['YearBuilt'].mode()
landsize_mode = melbourne['Landsize'].mode()
buildingarea_mode = melbourne['BuildingArea'].mode()
councilarea_mode = melbourne['CouncilArea'].mode()
lattitude_mode = melbourne['Lattitude'].mode()
longitude_mode = melbourne['Longtitude'].mode()
region_mode = melbourne['Regionname'].mode()
property_mode = melbourne['Propertycount'].mode()
```

```python
#Floor = q1(25%)-1.5*ICQ
#Cap= q2(75%)+1.5*ICQ
```

```python
Q1=n_melbourne['Rooms'].quantile(0.25)
Q3=n_melbourne['Rooms'].quantile(0.75)
IQR=Q3-Q1
Floor = Q1-1.5*IQR
Cap= Q3+1.5*IQR
print(Floor)
print(Cap)
n_melbourne1=n_melbourne[(n_melbourne['Rooms']>=Floor) & (n_melbourne['Rooms']<=Cap)]
sns.boxplot(n_melbourne1['Rooms'])
```
